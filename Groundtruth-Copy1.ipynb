```python
import networkx as nx

lines = []
with open('/Users/rheaverma/Downloads/dummy-data/small_simulation/save/graph/graph.1.txt') as f:
    lines = f.readlines()

nodes = []
edges = []

count = 0

length = len(lines)
for line in lines:
    count += 1
    print(line)
    if count == 2:
        nodes = line.strip().split(';')
        print(nodes)
    if count > 4 and count < (length):
        temp = []
        temp = line.strip().split(' ')
        
        edges.append((temp[1], temp[3]))
        print(edges)
        
```

```python
G = nx.DiGraph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
import matplotlib.pyplot as plt
nx.draw(G, with_labels=True)
plt.show()
```
